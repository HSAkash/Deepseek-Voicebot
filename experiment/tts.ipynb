```python
from gtts import gTTS
from pyprojroot import here

fileName = here("data/audio/temp_audio.mp3")

def play_text_to_speech(text, fileName, language='en', slow=False):
    tts = gTTS(text=text, lang=language, slow=slow)
    tts.save(fileName)
```

```python
play_text_to_speech("Hello, this is a test", fileName)
```
